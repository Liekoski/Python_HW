```python
'''____________________________Задание 0____________________________'''

import pprint

results = {'Иванова': {'Макулатура': 3, 'Металлолом': 1, 'Морковь': 4},
           'Петров': {'Макулатура': 10, 'Металлолом': 12, 'Морковь': 2},
           'Сидоров': {'Макулатура': 1, 'Металлолом': 1, 'Морковь': 3},
           'Алексеева': {'Макулатура': 1, 'Металлолом': 2, 'Морковь': 8},
           'Джонсон': {'Макулатура': 23, 'Металлолом': 5, 'Морковь': 2}}


#pprint.pprint(results['Петров']['Морковь'])

f = open('NewResults0.txt', encoding='cp1251')

for line in f:
    items = line.split()
    name = items[0] #имя
    collect = items[1]   #что собирали
    weight = int(items[2]) #сколько собрали
    results[name][collect] += weight

f.close()

pprint.pprint(results)
```

```python
'''____________________________Задание 1____________________________

 Ситуация усложняется.
Написать программу, которая принимает текстовый файл NewResults1.txt,
извлекает из него новые результаты и добавляет их в таблицу к имеющимся данным.
Формат менее удобный: вначале указывается материал, а затем на следующих строчках — кто сколько его собрал:

Морковь

Иванова - 3

Петров - 4'''

f = open('NewResults1.txt', encoding='cp1251')

for line in f:
  if '-' not in line:
    collect = line.strip()
  else:
    items = line.split(' - ')
    name = items[0]
    weight = int(items[1])
    results[name][collect] += weight

f.close()

pprint.pprint(results)
```

```python
'''____________________________Задание 2____________________________


Ситуация усложняется. Пионеры начали сдавать все подряд.
Необходимо подправить программу, чтобы она импортировала NewResults2.txt и добавляла новые собранные единицы.
Здесь кроме имеющихся моркови, макулатуры и металлолома может собираться что угодно. Например, ящики.'''


f = open('NewResults2.txt', encoding='cp1251')

for line in f:
  #print(line)
  if '-' not in line:
    collect = line.strip()
  else:
    items = line.split()
    name = items[0].strip()
    weight = int(items[2])
    if collect in results[name]: #проверяем, есть ли данный элемент
      results[name][collect] += weight
    else:
      results[name][collect] = weight

f.close()
pprint.pprint(results)
```

```python
'''____________________________Задание 3____________________________


Ситуация еще усложняется. Местные жители начали помогать пионерам, да и сами пионеры продолжают сдавать все подряд.
Необходимо усовершенствовать программу, чтобы она импортировала NewResults3.txt, добавляла новых людей и новые собранные единицы.'''

f = open('NewResults3.txt', encoding='cp1251')

for line in f:
  #print(line)
  if '-' not in line:
    collect = line.strip()
  else:
    items = line.split()
    name = items[0].strip()
    if name not in results:
      #continue - чтобы пропустить и пойти дальше
      results[name] = {}
    weight = int(items[2])
    if collect in results[name]: #проверяем, есть ли данный элемент
      results[name][collect] += weight
    else:
      results[name][collect] = weight

f.close()
pprint.pprint(results)
```
