# **CSV — современный способ хранить данные**

1) Данные обычно хранятся в файлах. Мы с вами работали с данными в формате txt, но это явно не самый удобный способ хранения. Сегодня рассмотрим еще один формат — csv.

В csv можно импортировать данные из Excel. Его удобно просматривать с помощью Excel.

Его название csv — сокращение от comma-separated values. Данные в csv разделяются запятыми (чаще всего), поэтому он и называется comma-separated values. Но в нашем примере, взятом из жизни, разделитель — точка с запятой.

5) Выведем еще три строчки.
ЭКСКУРС: После того, как мы вывели первую строчку, указатель оказался в начале второй строки, поэтому компьютер начнет обработку (вывод) со второй строки. Это важно помнить. Если вы работали с файлом, то указатель может оказаться где-то в середине или в конце, и начало файла может не попасть в обработку.


```python
for i in range(0, 3):
    test_line = all_crimes.readline()
    print(test_line)
```

Рассмотрите выведенные строчки  внимательнее. Сравнивая их с заголовками, можно понять, что это дата-время преступления, адрес, полицейский округ, потом beat (район  -- основное подразделение округа в ряде южных штатов США), затем непонятный затем crimedescr — crime description — описание преступления.

Полицейские округа в Скараменто названы цифрами: от 1 до 6. Эта информация нам пригодится.

ВОПРОС: почему при выводе данные разделены пустыми строчками?

Потому что в конце каждой сточки в файле стоит «переход на новую строку» ‘\n’. Из-за него после вывода начинается новая строка. А очередной print опять начинается с новой строки. Поэтому получается две «новых строки».



6) Давайте поправим вывод: пусть выводятся для трех первых строчек только время преступления  и округ.

Как вывести время преступления? Оно в строке данных идет самым первым.
Ответ: разбить строку по знаку «точка с запятой» и вывести нужный элемент (в нашем случае — нулевой).

Как нам разбить строку по запятой? Все, конечно, помнят, что есть метод split, который разбивает строку по разделителям и создает список. В нашем случае разделитель — точка с запятой. Напомню этот метод: items = test_line.split(';')


```python
items = test_line.split(',') #сплит создаёт список
print(f'Время: {items[0]}, округ: {items[2]}')
```
