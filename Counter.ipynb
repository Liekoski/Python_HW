Теперь посчитаем ягоды и грибы отдельно

```python
with open('berries_and_mushrooms.txt', 'r', encoding='cp1251') as f:
  for line in f:
    items = line.split()
    person = items[1]
    weight = int(items[2])
    taken = items[0]


    for take in taken:
      if taken == 'Ягоды:':
        c_berries[person] += weight

      if taken == 'Грибы:':
        c_mushrooms[person] += weight
```

```python
c_berries
```

```python
c_mushrooms
```

```python
print('Ягоды:\n')
for pair1 in c_berries.most_common():
  print(pair1)

print('\nГрибы:\n')
for pair2 in c_mushrooms.most_common():
  print(pair2)
```
