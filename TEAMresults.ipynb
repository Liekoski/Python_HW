Открыть файл, прочитать по строчке, нужно собрать имена, собрать total

0) Создать список имен всех участниц (без повторений).

1) Посчитать, сколько всего было собрано баллов.

2) В том же цикле посчитать, сколько баллов собрали команды, в которых была Лена.

Подсказка: для этого надо проверять, есть ли Лена в списке, который получился после разбиения строки на четыре элемента.

3) В том же цикле посчитать, сколько собрали команды, в которых была Надя.

4) В том же цикле посчитать средний балл для Лены и для Нади.

Подсказка: для этого надо разделись общий бал из пункта 2 на количество участий Лены.
И аналогичным образом посчитать для Нади (общий бал из пункта 2 на количество участий Нади).

5) Посчитать, сколько раз Лена и Надя выступали в одной команде.

```python
total = 0
names = set()
total_with_Lena = 0
total_with_Nadya = 0
items_with_Lena = 0
items_with_Nadya = 0
med_Lena = 0
med_Nadya = 0
items_with_Lena_and_Nadya = 0
```

```python
items_with_Lena_and_Nadya
```

```python
with open('results.txt', encoding='windows-1251') as f:
  for line in f:
    items = line.split()
    weight = int(items[-1])
    names.update(items[0:3])
    total = total + weight
    if 'Лена' in items:
      total_with_Lena += weight
      items_with_Lena += 1

    if 'Надя' in items:
      total_with_Nadya += weight
      items_with_Nadya += 1

    if 'Лена' and 'Надя' in items: #почему они играли вместе 11 раз? Если в документе видно, что они играли вместе всего 5 раз
      items_with_Lena_and_Nadya += 1


print('Лена поучаствовала:', items_with_Lena, 'раз')
print('Надя поучаствовала:', items_with_Nadya, 'раз')

med_Lena = total_with_Lena / items_with_Lena
med_Nadya = total_with_Nadya / items_with_Nadya

print('Среднее у Лены:', med_Lena)
print('Среднее у Нади:', med_Nadya)

print('Надя и Лена играли вместе:', items_with_Lena_and_Nadya)
```
