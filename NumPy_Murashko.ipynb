# **Задача 3**

```python
'''Создайте двумерный массив 8x8 и заполните его нулями и единицами в шахматном
порядке.'''


arr7 = np.zeros((8, 8)) #сначала создаём массив с нулями
for i in range(8):
  for j in range(8):
    if (i + j) % 2 == 0: #тут мы смотрим, если сумма индексов делится без остатка на два, то оставляем 0
      arr7[i, j] = 0
    else:
      arr7[i, j] = 1
print(arr7)

#наоборот
arr8 = np.zeros((8, 8)) #сначала создаём массив с нулями
for i in range(8):
  for j in range(8):
    if (i + j) % 2 == 0: #тут мы смотрим, если сумма индексов делится без остатка на два, то оставляем 0
      arr8[i, j] = 1
    else:
      arr8[i, j] = 0
print('\n\n', arr8)
```

# **Задание 4**

```python
'''Создайте массив 10х10, заполненный нулями (типа integer, без плавающей точки).
Затем заполните его пятерками ниже диагонали'''

arr9 = np.zeros((10, 10), dtype=int)
for i in range(10):
    for j in range(i + 1): #с каждым шагом увеличивается на 1
        arr9[i, j] = 5
print(arr9)
```
