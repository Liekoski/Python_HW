```python
import random
```

```python
examples_hello = ['привет', 'здравствуйте']
examples_bye = ['пока', 'досвидания']

responses_hello = ['Ну здравствуй, Владик...)', 'Шуршит перекатиполе на Диком Западе', 'Hello there!']
responses_bye = ['Уже уходишь? Ипугался???', 'Эх...', 'Покаааааааааа!!!!!!!!!!!!!!!!!']

failure_answ = ['Ну зачем так... пошло?!', 'Умные мысли сопровождали его, но он был быстрее', 'Я не понимаю!!! Ничего не понимаю! Ни-че-го!!']


def get_intent(question):
  question = question.lower() #переводим в маленький регистр, проверка пройдена
  if question in examples_hello:
    intent = 'hello'
  elif question in examples_bye:
    intent = 'bye'
  else:
    intent = None
  return(intent)

def bot(question):
  intent = get_intent(question)
  if intent == 'hello':
    answer = random.choice(responses_hello)
  elif intent == 'bye':
    answer = random.choice(responses_bye)
  else:
    answer = random.choice(failure_answ)
  #answer = question
  return(answer)

print('Бот готов к работе!')

while True:
  question = input('Вы: ')
  if question == "стоп":
    break
  answer = bot(question)
  print('Бот: ', answer)
```
